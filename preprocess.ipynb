```python
import pandas as pd
```

```python
raw_path = './raw_csv'
cleaned_path = './cleaned_csv'
```

### พิกัด

```python
latlong = pd.read_csv('{}/latlong.csv'.format(raw_path))
latlong.head()
```

```python
province = latlong['CHANGWAT_T'].unique()
province = [p.replace('จ. ', '') for p in province]
province
```

```python
latlong['province'] = latlong['CHANGWAT_T'].apply(lambda x: x.replace('จ. ', ''))
```

```python
district = latlong[latlong['AMPHOE_T'].notnull()]['AMPHOE_T'].unique()
district = [d.replace('อ. ', '') for d in district]
district
```

```python
latlong['district'] = latlong['AMPHOE_T'].apply(lambda x: x.replace('อ. ', '') if isinstance(x, str) else x)
```

```python
latlong.to_csv('{}/latlong.csv'.format(cleaned_path), index=False)
```

### ข้าว ฯลฯ

```python
rice = pd.read_csv('{}/rice.csv'.format(raw_path))
rice.head()
```

```python
rice_by_province = rice.loc[rice['พื้นที่'].isin(province)]
rice_by_province
```

```python
rice_by_province.to_csv('{}/rice_by_province.csv'.format(cleaned_path), index=False)
```

```python
rice_by_district = rice.loc[rice['พื้นที่'].isin(district)]
rice_by_district
```

```python
rice_by_district.to_csv('{}/rice_by_district.csv'.format(cleaned_path), index=False)
```
